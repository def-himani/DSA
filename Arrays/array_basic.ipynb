# Array operations

```python
A = [1,2,4,6,10]
print(A)
```

```python
# O(1)
A.append(9)
A
```

```python
# O(n)
A.insert(1,111)
A
```

```python
# O(1)
A.pop()
```

```python
A
```

```python
# Modify - O(1)
A[2] = 0
A
```

```python
# Access an element
A[4]
```

```python
# Checking if a val is in A - O(n)
print(7 in A)
```
